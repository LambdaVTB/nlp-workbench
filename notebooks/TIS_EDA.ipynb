```python
import pandas as pd
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=True)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

from corus import load_lenta2
```

```python
# Read csv file
path = 'S:\Workspace\MORETECH\nlp-workbench\lenta-ru-news-tail-15000.csv'
def load_records_lenta(path: str, sample_count: int, is_compressed: bool = True) -> pd.DataFrame:
    if is_compressed:
        records = load_lenta2(path)
    else:
        df = pd.read_csv(path, sep=',', encoding='utf-8', engine='pyarrow')
        df.columns = ['url', 'title', 'text', 'topic', 'tags', 'date']
        df['source'] = 'lenta'
        return df

    df = pd.DataFrame(columns=['title', 'text', 'url', 'topic', 'tags', 'date'])
    for i, record in tqdm(enumerate(records), total=sample_count):
        df.loc[i] = [record.title, record.text, record.url, record.topic, record.tags, record.date]
        if i > sample_count:
            break
    df['source'] = 'lenta'
    return df

df = load_records_lenta(path, 10000, False)
df.head()
```

```python
df_tail = df.tail(1000)
df_tail.tail()
```

```python
df_tail.to_csv('S:\Workspace\MORETECH\nlp-workbench\lenta-news-sample-1000.csv', index=False)
```
